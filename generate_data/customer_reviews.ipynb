```python
products = [
    ('iPhone 15 Pro', 'Electronics', 41900),
    ('Samsung S24 Ultra', 'Electronics', 45900),
    ('Sony WH-1000XM5', 'Accessories', 11900),
    ('MacBook Air M2', 'Computers', 34900),
    ('Logitech MX Master 3S', 'Accessories', 3900),
    ('Dell XPS 13', 'Computers', 55000),
]
products
```

```python
import numpy as np
extended_reviews = []
for _ in range(50):
    base_review = random.choice(reviews_data)
    sentiment = "Positive" if "ชอบ" in base_review or "ดี" in base_review or "คุ้ม" in base_review else "Negative"
    extended_reviews.append({
        "review_id": f"REV-{random.randint(1000, 9999)}",
        "product_name": random.choice([p[0] for p in products]),
        "comment": base_review,
        "sentiment": sentiment
    })
```

```python
df_reviews = pd.DataFrame(extended_reviews)
df_reviews.head()
```

```python
df_reviews.to_csv("../data/customer_reviews.csv", index=False)
```
