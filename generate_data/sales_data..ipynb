```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
products = [
    ('iPhone 15 Pro', 'Electronics', 41900),
    ('Samsung S24 Ultra', 'Electronics', 45900),
    ('Sony WH-1000XM5', 'Accessories', 11900),
    ('MacBook Air M2', 'Computers', 34900),
    ('Logitech MX Master 3S', 'Accessories', 3900),
    ('Dell XPS 13', 'Computers', 55000),
]
products
```

```python
data = []

for i in range(500):
    prod = random.choice(products)
    date = datetime(2024, 1, 1) + timedelta(days=random.randint(0, 90))
    qty = random.randint(1, 3)
    
    region = ["Bangkok", "Chiang Mai", "Phuket", "Khon Kaen"]
    
    data.append({
        "transaction_id": f"TRX-{1000+i}",
        "date": date.strftime("%Y-%m-%d"),
        "product_name": prod[0],
        "category": prod[1],
        "unit_price": prod[2],
        "quantity": qty,
        "total_amount": prod[2] * qty,
        "region": random.choice(region)
    })
```

```python
df_sales = pd.DataFrame(data)
df_sales.head()
```

```python
df_sales.to_csv("../data/sales_data.csv", index=False)
```
